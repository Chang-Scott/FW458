```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Problem 1
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot data with dashed line style and blue color
plt.plot(x, y, linestyle='--', color='blue')

# Add title
plt.title('Sine Wave')
# Add axis titles
plt.xlabel('X')
plt.ylabel('sin(X)')
```

```python
# Problem 2
tips = sns.load_dataset("tips") # Get tips data
# Plot tips data using displot of kind histogram and with kde set to True t overlay KDE
g = sns.displot(data = tips, x = 'total_bill', bins=20, color="blue", kind = 'hist', kde = True) # Use bins 20
g.set_axis_labels("Total Bill ($)", "Frequency")
plt.title("Total Bill Distribution")
```

```python
# Problem 3
tips = sns.load_dataset("tips")
# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2)

# Here we define that the axis we are plotting on is the first column for boxplot and second column for violinplot
# set red for Lunch and blue for Dinner
sns.boxplot(data = tips, y = 'tip',  x = 'time', hue = 'time', ax=axes[0], palette=['red', 'blue'])
sns.violinplot(data = tips, y = 'tip', x = 'time', hue = 'time', ax=axes[1], palette=['red', 'blue'])

# Now let's set title and axes
plt.suptitle('Plot of Tips')
axes[0].set_xlabel("Box Plot of Tips")
axes[0].set_ylabel("Tips")
axes[1].set_xlabel("Violin Plot of Tips")
axes[1].set_ylabel("Tips")
```

```python
# Problem 4

# Create a scatter plot with regression lines
g = sns.lmplot(data=tips, y='total_bill', x='tip', hue='smoker', palette=['red', 'blue']) # Use red for smokers and blue for non-smokers
g.set_axis_labels("Tip ($)", "Total Bill ($)") # set axis labels
g.add_legend(frameon = True) # Set legend
plt.title("Total Bill vs. Tip") # Set title
```

```python
# Problem 5
 # Get the correlation matrix using corr function
correlation_matrix = tips.corr(numeric_only = True)
# Ended up setting numeric only to True since it was throwing ValueError since tips has string columns

# Create the heatmap using sns.heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') # Use coolwarm, which is a diverging colormap
# Set annotate to True so we get annotations
```

```python
# Problem 6
# Make pairwise plots
g = sns.pairplot(data=tips, hue='time', palette=['red', 'blue']) # Use red for lunch and blue for dinner
# Make diagonal plots with KDE
g.map_diag(sns.kdeplot)
```
